```python
import pandas as pd
import numpy as np

import requests
from bs4 import BeautifulSoup
import urllib.request as ur
```

#### Alabama
Executive orders of the State Health Officer

```python
# 
url = 'https://www.alabamapublichealth.gov/legal/orders.html#adph'
```

```python
# get html from url
r = requests.get(url)
r.status_code
```

```python
# create BeautifulSoup object
soup = BeautifulSoup(r.content, 'lxml')
```

```python
# show html
soup.extract()
```

```python
# find and select the ul that contains all href to Health Department executive orders
ul = soup.find_all('ul')
ul[2]
```

```python
# find all hrefs
all_a = ul[2].find_all('a')
all_a
```

```python
# loop through 'a's and extract partial_urls

href_list = []

for a in all_a:
    # select only those partial urls which start with 'assets'
    if a.attrs['href'].split('/')[0] == 'assets':
        if 'spanish' not in a.attrs['href'].split('-'):
            href_list.append(a.attrs['href'])
    
```

```python
# create urls from partial urls
# download pdf files 
# save them in folder Alabama

base_url = 'https://www.alabamapublichealth.gov/legal/'
counter = 0
for partial in href_list:
    counter += 1
    pdf_url = base_url + partial
    ur.urlretrieve(pdf_url, './data/Alabama/alabama'+ str(counter) + '.pdf')
    
# source for retrieving pdf files from internet: 
# https://medium.com/@umerfarooq_26378/python-for-pdf-ef0fac2808b0
```

#### Alaska

Health Mandates of the Governor

```python
# base url
url = 'https://covid19.alaska.gov/health-mandates/'

# request url
# check if url is working
res = requests.get(url)
res.status_code
```

```python
# instantiate BeautifulSoup
soup_alaska = BeautifulSoup(res.content, 'lxml')

# extract html for analysis
soup_alaska.extract()
```

```python
# get text from current documents
# instantiate empty list of documents
mandate_list = []


counter = 5690
for i in range(1, 11):
    counter += 1
    if counter <= 5699:
        counter = counter
    elif counter > 5699:
        counter = 56910
   
    # instantiate empty string
    text = ''
    
    # find all instances of 'p' in div where the id_number is the counter
    for p in soup_alaska.find('div', {'id': 'elementor-tab-content-' + str(counter)}).find_all('p')[2:]:
        # extract all text from p
        text += p.text
        
        document = {
            'title': soup_alaska.find('div', {'id': 'elementor-tab-title-' + str(counter)}).\
                     find('a', {'class': 'elementor-accordion-title'}).text,
            'text': text
        }
   
    
    # create list of all documents
    mandate_list.append(document)
```

```python
# extract text from expired or superseded health mandates



counter = 9820
for i in range(0, 7):
    counter += 1
   
    # instantiate empty string
    text = ''
    
    # find all instances of 'p' in div where the id_number is the counter
    for p in soup_alaska.find('div', {'id': 'elementor-tab-content-' + str(counter)}).find_all('p')[2:]:
        # extract all text from p
        text += p.text
        
        document = {
            'title': soup_alaska.find('div', {'id': 'elementor-tab-title-' + str(counter)}).\
                     find('a', {'class': 'elementor-accordion-title'}).text,
            'text': text
        }
   
    
    # create list of all documents
    mandate_list.append(document)
```

```python
data = pd.DataFrame(mandate_list)
```

```python
# export data as csv

data.to_csv('./data/CSV_files/alaska.csv', index=False)
```
